# Boston House Price

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
boston = load_boston()
boston
```

```python
X=boston.data
y=boston.target
```

```python
reg=LinearRegression()
lasso_reg=Lasso(alpha=10)
ridge_reg=Ridge(alpha=10)

reg.fit(X,y)
lasso_reg.fit(X,y)
ridge_reg.fit(X,y)
```

```python
x12=[]
for i in range(506):
    x12.append(X[i,12])
X12=np.array(x12).reshape(-1,1)
y=np.array(y).reshape(-1,1)
```

```python
reg.fit(X12,y)
lasso_reg.fit(X12,y)
ridge_reg.fit(X12,y)

predict_reg = reg.predict(X12)
predict_lasso = lasso_reg.predict(X12)
predict_ridge = ridge_reg.predict(X12)
```

```python
plt.plot(X12,y,'.')
plt.plot(X12,reg.predict(X12), color="red", label="Reg.")
plt.legend()

plt.title("Boston Linear Regression")
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

```python
plt.plot(X12,y,'.')
plt.plot(X12,lasso_reg.predict(X12), color="green", label="Lasso Reg.")
plt.plot(X12,ridge_reg.predict(X12), color="orange", label="Ridge Reg.")

plt.title("Boston Linear Regression")
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

```python
print(" Slope / Score")
print("- Reg. : ",round(reg.coef_[0,0],3),'/',round(reg.score(X12,y),3))
print("- Lasso Reg. : ",round(lasso_reg.coef_[0],3),'/',round(lasso_reg.score(X12,y),3))
print("- Ridge Reg. : ",round(ridge_reg.coef_[0,0],3),'/',round(ridge_reg.score(X12,y),3))
```
