# Diabetes

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dia = load_diabetes()
print(dia.DESCR)
```

```python
X=dia.data
y=dia.target
```

```python
x1=[]
for i in range(442):
    x1.append(X[i,2])
X1=np.array(x1).reshape(-1,1)
y=np.array(y).reshape(-1,1)
```

```python
reg=LinearRegression()
lasso_reg=Lasso(alpha=0.1)
ridge_reg=Ridge(alpha=0.1)

reg.fit(X1,y)
lasso_reg.fit(X1,y)
ridge_reg.fit(X1,y)

predict_reg = reg.predict(X1)
predict_lasso = lasso_reg.predict(X1)
predict_ridge = ridge_reg.predict(X1)
```

```python
plt.plot(X1,y,'.')
plt.plot(X1,reg.predict(X1), color="red", label="Reg.")
plt.legend()

plt.title("Diabetes Linear Regression")
plt.xlabel('BMI')
plt.ylabel('Dia_progress')
```

```python
plt.plot(X1,y,'.')
plt.plot(X1,lasso_reg.predict(X1), color="green", label="Lasso Reg.")
plt.plot(X1,ridge_reg.predict(X1), color="orange", label="Ridge Reg.")
plt.legend()

plt.title("Diabetes Linear Regression")
plt.xlabel('BMI')
plt.ylabel('Dia_progress')
```

```python
print("▶ Slope / Score")
print("- Reg. : ",round(reg.coef_[0,0],3),'/',round(reg.score(X1,y),3))
print("- Lasso Reg. : ",round(lasso_reg.coef_[0],3),'/',round(lasso_reg.score(X1,y),3))
print("- Ridge Reg. : ",round(ridge_reg.coef_[0,0],3),'/',round(ridge_reg.score(X1,y),3))
```
